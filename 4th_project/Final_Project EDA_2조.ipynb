```python
# 라이브러리 불러오기
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data1 = pd.read_csv('/content/drive/MyDrive/lsda02/data/component1_droped.csv')
data2 = pd.read_csv('/content/drive/MyDrive/lsda02/data/component2.csv')
data3 = pd.read_csv('/content/drive/MyDrive/lsda02/data/component3.csv')
data4 = pd.read_csv('/content/drive/MyDrive/lsda02/data/component4.csv')
```

### Zn-P 산점도

```python
import matplotlib.pyplot as plt

# 예시 데이터
# data4 = pd.read_csv('your_dataset.csv')  # 데이터가 이미 있는 경우

# Y_LABEL 값을 색상으로 매핑
colors = data1['Y_LABEL'].map({0: 'skyblue', 1: 'orange'})

plt.figure(figsize=(10, 6))
scatter = plt.scatter(data1['P'], data1['ZN'], c=colors, alpha=0.5)
plt.title('Scatter of Component1 Zn - P')
plt.xlabel('P')
plt.ylabel('ZN')
plt.grid(True)

# 범례 추가
legend_labels = {0: 'Class 0', 1: 'Class 1'}
handles = [plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='skyblue', markersize=10, label='Class 0'),
           plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='orange', markersize=10, label='Class 1')]

plt.legend(handles=handles, title="Y_LABEL")

plt.show()
```
